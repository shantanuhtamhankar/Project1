```python
# Let's Import all the necessory Librarys for our project
# Such as numpy, pandas, matplotlib, seaborn
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Now let's import the file in which data is saved
data = pd.read_csv("Iris.csv")
```

```python
# view first 5 rows of our data
data.head()
```

```python
# To check total count, mean, median of every variable given to us
data.describe()
```

```python
# To chech data type of every characteristic (variable)
data.info()
```

```python
# let's find out that is there any missing value
data.isnull().sum()
```

```python
data.describe()
```

```python
# To count every unique value of veriable having object as a data type
print(data.Species.value_counts())
```

```python
# To convert such object data type characteristics to int data type
data.replace({"Species":{"Iris-setosa":0,"Iris-versicolor":1,"Iris-virginica":2}},inplace=True)
```

```python
data.head()
```

```python
# To draw correlation matrics for visualising the relationship od every variable with each other
cm=data.corr() 
sns.heatmap(cm,annot=True,cmap='coolwarm') 
plt.title("Correlation Matrix") 
plt.show()
```

# By observing above correlation matrics we can say that out of all independent variables there is major impact of PetalwidthCm on Species (dependent variable)

```python
# To split the data into x and y
x = data.iloc[:,1:5]
y = data.iloc[:,5]
```

```python
# Let's check sepration of x and y is succeed or not
x.head()
```

```python
y.head()
```

```python
# Now let us split the data horizontally in training data and testing data by trsin_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, random_state=34, test_size=0.2)
```

# Support Vector regression

```python
# TO apply/fit Support Vector regression model on our data
from sklearn.svm import SVC
model_svm = SVC()
model_svm.fit(x_train,y_train)
```

```python
# let's predict the value of dependent variable with the help of independent testing data
SVR_y_pred = model_svm.predict(x_test)
```

```python
plt.figure(figsize=(5, 3))
plt.scatter(y_test, SVR_y_pred, alpha=0.7)
plt.title('Actual vs Predicted Sales', fontsize=10)
plt.xlabel('Actual Sales', fontsize=8)
plt.ylabel('Predicted Sales', fontsize=8)
plt.grid(True)
plt.show()
```

```python
# Now to check fitness of our model with the help of R2 score
from sklearn.metrics import r2_score
r2 = r2_score(y_test, SVR_y_pred)
print("R-squared Score:", r2)
```

# Logistic Regression

```python
# TO apply/fit Logistic Regression model on our data
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit (x_train, y_train)
```

```python
# let's predict the value of dependent variable with the help of independent testing data
LR_y_pred = model_LR.predict(x_test)
```

```python
plt.figure(figsize=(5, 3))
plt.scatter(y_test, LR_y_pred, alpha=0.7)
plt.title('Actual vs Predicted Value', fontsize=10)
plt.xlabel('Actual Value', fontsize=8)
plt.ylabel('Predicted Value', fontsize=8)
plt.grid(True)
plt.show()
```

```python
# Now to check fitness of our model with the help of R2 score
from sklearn.metrics import r2_score
r2 = r2_score(y_test, LR_y_pred)
print("R-squared Score:", r2)
```

# Decision Tree Classification

```python
# TO apply/fit Decision Tree Classification model on our data
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
model_tree.fit(x_train, y_train)
```

```python
# let's predict the value of dependent variable with the help of independent testing data
tree_y_pred = model_tree.predict(x_test)
```

```python
# To visualise the relation between original sales (y_test) and predicted sales (tree_y_pred)
plt.figure(figsize=(5, 3))
plt.scatter(y_test, tree_y_pred, alpha=0.7)
plt.title('Actual vs Predicted Sales', fontsize=10)
plt.xlabel('Actual Sales', fontsize=8)
plt.ylabel('Predicted Sales', fontsize=8)
plt.grid(True)
plt.show()
```

```python
# Now to check fitness of our model with the help of R2 score
from sklearn.metrics import r2_score
r2 = r2_score(y_test, tree_y_pred)
print("R-squared Score:", r2)
```
